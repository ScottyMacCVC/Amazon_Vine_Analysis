```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# import and read csv file
df = pd.read_csv('Resources/vine_table.csv', index_col=0)
df
```

```python
# filter dataframe to show >= 20 votes
filter_df = df[df['total_votes'] >= 20]
filter_df
```

```python
# calculate variable for helpful votes as a percent of all votes
filter_df['%_helpful'] = filter_df.helpful_votes / filter_df.total_votes

#filter table to only show rows with % positive greater than 50%
helpful_df = filter_df[filter_df['%_helpful'] >= .50]
```

```python
# check filter
helpful_df['%_helpful'].min()
```

```python
# filter step2 df to only show rows with a Y value for a Vine review
helpful_df_filtered_y = helpful_df[helpful_df['vine'] == 'Y']
```

```python
# filter dataframe for vine = N
helpful_df_filtered_n = helpful_df[helpful_df['vine'] == 'N']
```

```python
# calculate number of reviews, number of 5 star reviews, percent of 5 star reviews for vine = y
total_count_y = len(helpful_df_filtered_y.index)
five_count_y_df =  helpful_df_filtered_y[helpful_df_filtered_y['star_rating'] == 5.0]
five_count_y  =len(five_count_y_df)
five_star_percent = five_count_y / total_count_y
print('Summary Data: Vine review')
print('--------------------------')
print(f'five star count: {five_count_y}')
print(f'total count: {total_count_y}')
print(f'five star percent: {five_star_percent:.1%}')
print('--------------------------')
```

```python
# calculate number of reviews, number of 5 star reviews, percent of 5 star reviews for vine = n
total_count_n = len(helpful_df_filtered_n.index)
five_count_n_df =  helpful_df_filtered_n[helpful_df_filtered_n['star_rating'] == 5.0]
five_count_n  =len(five_count_n_df)
five_star_percent = five_count_n / total_count_n
print('Summary Data: No Vine review')
print('--------------------------')
print(f'five star count: {five_count_n}')
print(f'total count: {total_count_n}')
print(f'five star percent: {five_star_percent:.1%}')
print('--------------------------')
```
